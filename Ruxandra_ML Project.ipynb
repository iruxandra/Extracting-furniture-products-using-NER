```python
#Training and testing set for matches on simple text - similar to what's happening in line 63

def parse_train_data (doc4):
    detections = [(doc4[start:end].start_char, doc4[start:end].end_char, ' FURNITURE PRODUCT') for idx, start, end in matcher(doc4)]
    return (doc4.text, {'entities': detections})

parse_train_data(nlp("Today I was not feeling well and decided to stay in bed on my soft mattress. Eventually I had to get up and go to my desk to do some work."))
```

# Data sets for ML model

```python
#Reminder: label_df = df4 but labelled with 0 and 1 for incorrect and correct matches, respectively

#Selecting 80% of the data for training 

#TRAINING set:

slct_label_df =  pd.read_csv('Crawled pages.txt',nrows = 120, delimiter="\t", usecols=['Label','Product'])
#removing new lines ('\n') in text file
slct_label_df = slct_label_df.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=["",""], regex=True, inplace=False)
products = slct_label_df.loc[lambda d: d['Label'] == 1]['Product']
TRAIN_DATA = [parse_train_data(d) for d in nlp.pipe(products) if len(matcher(d)) > 1]


#Note for line : for d in nlp.pipe(products) if len(matcher(d)) ==1
#What it does: when the matcher finds one product, it will let that be our training data

#VALIDATION set:
vld_label_df =  pd.read_csv('Crawled pages.txt',skiprows = [i for i in range(1,121)], delimiter="\t", usecols=['Label','Product'])
 #removing new lines ('\n') in text file
vld_label_df = vld_label_df.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=["",""], regex=True, inplace=False)
validation = vld_label_df.loc[lambda d: d['Label'] == 1]['Product']
VALID_DATA = [parse_train_data(d) for d in nlp.pipe(validation) if len(matcher(d)) >1 ]

#TEST data set:
TEST_DATA =  pd.read_csv('Crawled pages.txt',skiprows = [i for i in range(1,121)], delimiter="\t", usecols=['Label','Product'])
#removing new lines ('\n') in text file
TEST_DATA = TEST_DATA.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=["",""], regex=True, inplace=False)


#TRAIN_DATA
#VALID_DATA
TEST_DATA
```

# Training Loop

- following the spacy documentation for this step
- link is here: https://spacy.io/usage/training#training-simple-style
- (label_df) represents 'gold standard' annotated data

- The Example class ensures that spacy can rely on the one standardised format that's passed through the pipeline

- Technique to improve the learning results is to set a dropout rate: rate at which it will drop individual features

```python
from tqdm import tqdm #loading bar

nlp=spacy.blank("en") #load a new empty spacy model
db = DocBin()

for text, annot in tqdm(TRAIN_DATA):
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in annot["entities"]:
        span = doc.char_span(start, end, label = label, alignment_mode = "contract")
        if span is None:
            pass
        else:
            ents.append(span)
        try:
            doc.ents = ents
        except:
            db.add(doc)
            print(db, db.attrs)
            
#save the docbin object
db.to_disk("./train.spacy")
```

```python
#Validation set 

db = DocBin()

for text, annot in tqdm(VALID_DATA):
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in annot["entities"]:
        span = doc.char_span(start, end, label = label, alignment_mode = "contract")
        if span is None:
            pass
        else:
            ents.append(span)
        try:
            doc.ents = ents
        except:
            db.add(doc)
            print(db, db.attrs)
            
#save the docbin object
db.to_disk("./valid.spacy")
```
